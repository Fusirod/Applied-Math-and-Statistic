## 1. Giới thiệu
### 1.1 Hồi quy tuyến tính là 1 quá trình dựa vào dữ liệu đã có để đưa ra 1 hàm xác định cho dữ liệu, chi tiết như sau:

> **Bước 1:** Cho 1 mô hình dữ liệu được mô phỏng bằng phương trình: $ Y =  \theta_0 + \theta_1X_1 + \theta_2X_2 + ... + \theta_kX_k $ với $X_i$ là các đặc trưng của X

> **Bước 2:** Lập các ma trận $ X, Y, \theta $ trong đó:
- $ X = (X_1X_2..X_k) $ với $X_i$ là các cột chứa dữ liệu đặc trưng thứ i của X
- $ Y $ là ma trận cột chứa các kết quả của mô hình
- $ \hat{\theta} $ là ma trận cột chứa $ \theta_i $

> **Bước 3:** Sử dụng công thức `OLS (Ordinary Least Square)`: $ \hat{\theta} = (X^{T}.X)^{-1}.X^{T}.Y $

> **Bước 4:** Tính tổng bình phương sai số bằng công thức tính `MSE (Medium Square Errors)`: $ \frac{1}{n}\sum_{i=1}^{n}(y_i - ŷ_i)^{2} $ với $ y_i, ŷ_i $ lần lượt là kết quả có sẵn và kết quả của mô hình vừa tìm

### 1.2 Xích Markov là một mô hình toán học mô tả một quá trình ngẫu nhiên, trong đó trạng thái hiện tại chỉ phụ thuộc vào trạng thái ngay trước đó, chứ không phụ thuộc vào toàn bộ lịch sử trước đó.

> **Thành phần cơ bản:** 
- Tập trạng thái $S$ gồm các trạng thái của xích
- Ma trận chuyển trạng thái $P = (p_{ij})$ có kích thước bằng kích thước của tập trạng thái $S$, các phần tử $p_{ij}$ là xác suất chuyển từ trạng thái $S_j$ sang trạng thái $S_i$
- Vector phân phối xác suất $\pi$

> **Định nghĩa 1:** Chuỗi {$X_n$} được gọi là chuỗi Markov nếu nó thỏa mãn `điều kiện Markov`:

$$
P(X_{n+1} = s| X_0 = x_0, X_1 = x_1, ..., X_{n-1} = x_{n-1}),\forall n > 1, \forall x_i, s \in S 
$$

> **Định nghĩa 2:** Xích Markov {$X_n$} được gọi là `đồng nhất` khi

$$
P(X_{n+1} = i| X_n = j) = P(X_1 = i| X_0 = j), \forall n,i,j \geq 0
$$

- Ma trận chuyển trạng thái $P = (p_{ij})$ chứa các xác suất chuyển trạng thái: $p_{ij} = P(X_{n+1} = i| X_n = j)$
- Trong tài liệu này, các Xích Markov được coi là đồng nhất, với không gian trạng thái $S$ = {1,2,3,...,k}, ma trận chuyển trạng thái $P$ có kích thước $k\times k$ và phân phối xác xuất $\pi_n, \forall n \geq 0$ trong đó $\pi_n(i) = P(X_n = i), \forall i \in S$

> **Mệnh đề 3:** Cho 1 Xích Markov {$X_n$}, ta có:
$$
P(X_{m+n} = i| X_m = j) = P(X_n = i| X_0 = j), \forall m,n \geq 0, \forall i,j \in S
$$
- Ma trận $P^{(n)} = (p_{ij}^{(n)})$ với $p_{ij}^{(n)} = P(X_n = i| X_0 = j)$ được gọi là ma trận chuyển trạng thái n bước với $p_{ij}^{(n)}$ được gọi là xác suất chuyển tiếp n bước. Cụ thể, $P$ là ma trận chuyển tiếp 1 bước

> **Định lý 4 (Phương trình Chapman_Kolmogorov):** Cho 1 Xích Markov {$X_n$}, ta có:
$$
P(X_{n+1} = i| X_0 = j) = \sum_{l = 1}^k P(X_{n+1} = i| X_n = l) \dot P(X_n = 1|X_0 = j), \forall n \geq 0, \forall i,j \in S
$$
- Với $P(X_{n+1} = i|X_n = l) = P(X_1 = i|X_0 = l) = p_{il}$, phương trình trên có thể được thu gọn thành: $$p_{ij}^{(n+1)} = \sum_{l = 1}^k p_{il}.p_{il}^{(n)}$$

> **Hệ quả 5:** `Ma trận chuyển tiếp n bước` bằng lũy thừa bậc n của $P$
$$
P^{(n+1)} = P \times P^{(n)},\forall n \in N
$$

> **Hệ quả 6:** Cho 1 Xích Markov {$X_n$} với phân phối đầu $\pi_0$, ta có:
$$
\pi_n = P^n.\pi_0, \forall n \in N
$$
$$
\pi_n(i) = P(X_n = i), \forall i = 1,2,3,...,k
$$

> **Định nghĩa 7:** Cho 1 Xích Markov {$X_n$} có phân phối trạng thái $\pi_i, \forall i \geq 0$ và $\sum_{i=1}^k \pi(i) = 1$
- $\pi$ đươc gọi là `phân phối giới hạn` của {$X_n$} nếu $$\lim_{n \to \infty} P^n (i,j) = \pi_n, \forall i,j = 1,2,...,k$$
- $\pi$ được gọi là `phân phối dừng` của {$X_n$} nếu $$ \pi = P\pi$$

> **Định nghĩa 8:** Ma trận chuyển trạng thái $P$ của 1 Xích Markov là `chính quy` nếu tồn tại $n$ sao cho tất cả các phần tử của $P^n$ đều dương

> **Mệnh đề 9:** Cho 1 Xích Markov {$X_n$} có `ma trận chuyển trạng thái P chính quy`, khi đó `phân phối dừng có giá trị bằng với phân phối giới hạn`

> **Câu b:** Xây dựng mô hình đơn đặc trưng: $sprice = \theta_0 + \theta_1 .X_1$

```python
def transpose_matrix(matrix):
    so_hang = len(matrix)
    so_cot = len(matrix[0])

    transposed_matrix= [[0 for _ in range(so_hang)] for _ in range(so_cot)]

    for i in range(so_hang):
        for j in range(so_cot):
           transposed_matrix[j][i] = matrix[i][j]
      
    return transposed_matrix
    
def inverse_matrix(matrix):    
    n = len(matrix)
    if any(len(row) != n for row in matrix):
        raise ValueError("Lỗi: Ma trận đầu vào phải là ma trận vuông.")
    
    A = [row[:] for row in matrix]
    
    for i in range(n):
        identity_row = [1.0 if j == i else 0.0 for j in range(n)]
        A[i].extend(identity_row)

    for i in range(n):
        max_row = i
        for k in range(i + 1, n):
            if abs(A[k][i]) > abs(A[max_row][i]):
                max_row = k
        A[i], A[max_row] = A[max_row], A[i]

        pivot = A[i][i]
        if abs(pivot) < 1e-10:
            raise ValueError("Lỗi: Ma trận không khả nghịch (singular matrix).")

        for j in range(i, 2 * n):
            A[i][j] /= pivot

        for k in range(n):
            if k != i:
                factor = A[k][i]
                for j in range(i, 2 * n):
                    A[k][j] -= factor * A[i][j]
                    
    inverse_matrix = [row[n:] for row in A]

    return inverse_matrix

def matrix_multiply(A, B):
    rows_A = len(A)
    cols_A = len(A[0])
    rows_B = len(B)
    cols_B = len(B[0])

    if cols_A != rows_B:
        raise ValueError("Số cột của ma trận A phải bằng số hàng của ma trận B.")

    C = [[0 for _ in range(cols_B)] for _ in range(rows_A)]

    for i in range(rows_A):
        for j in range(cols_B):
            for k in range(cols_A):
                C[i][j] += A[i][k] * B[k][j]
    return C

def OLS(X,y):
    return matrix_multiply(inverse_matrix(matrix_multiply(transpose_matrix(X), X)), matrix_multiply(transpose_matrix(X), y))

def predict(X_b, beta):
    return matrix_multiply(X_b, beta)

def MSE(y_true, y_pred):
    rss = 0.0
    for i in range(len(y_true)):
        error = y_true[i][0] - y_pred[i][0]
        rss += error ** 2
    return rss/len(y_true)

def k_fold_cross_validation(data, headers, feature_col_names, target_col_name, k=100):
    feature_indices = [headers.index(name) for name in feature_col_names]
    target_index = headers.index(target_col_name)

    fold_size = len(data) // k
    total_mse = 0.0
    
    for i in range(k):
        start = i * fold_size
        end = start + fold_size if i != k - 1 else len(data)
        
        test_data = data[start:end]
        train_data = data[:start] + data[end:]

        X_train_data = [[row[j] for j in feature_indices] for row in train_data]
        y_train_data = [[row[target_index]] for row in train_data]
        X_train_b = [[1.0] + row for row in X_train_data]

        X_test_data = [[row[j] for j in feature_indices] for row in test_data]
        y_test_true = [[row[target_index]] for row in test_data]
        X_test_b = [[1.0] + row for row in X_test_data]

        try:
            beta = OLS(X_train_b, y_train_data)
            y_test_pred = predict(X_test_b, beta)
            fold_mse = MSE(y_test_true, y_test_pred)
            total_mse += fold_mse
        except ValueError:
            continue
            
    return total_mse / k

df = pd.read_csv('stockton4.csv')
df.columns = df.columns.str.strip()
headers = list(df.columns)
data = df.values.tolist()
feature_cols_b = [h for h in headers if h != 'sprice']
target_col_b = 'sprice'
k_folds = 100

best_feature_b = None
lowest_avg_mse_b = float('inf')
feature_mse_scores = {}

for feature in feature_cols_b:
    avg_mse = k_fold_cross_validation(data, headers, [feature], target_col_b, k=k_folds)
    feature_mse_scores[feature] = avg_mse
    
    if avg_mse < lowest_avg_mse_b:
        lowest_avg_mse_b = avg_mse
        best_feature_b = feature

print("\n--- Kết quả ---")
print(f"Đặc trưng đơn lẻ tốt nhất là: '{best_feature_b}'")
print(f"Với MSE trung bình qua {k_folds}-Fold CV là: {lowest_avg_mse_b:.4f}")

best_feature_index = headers.index(best_feature_b)
target_index = headers.index(target_col_b)

X_data_best = [[row[best_feature_index]] for row in data]
y_data_best = [[row[target_index]] for row in data]
X_b_best = [[1.0] + row for row in X_data_best]

beta_best = OLS(X_b_best, y_data_best)
beta_coeffs_best = [b[0] for b in beta_best]

print("\n--- PHƯƠNG TRÌNH HỒI QUY (1 ĐẶC TRƯNG TỐT NHẤT) ---")
equation_b = f'sprice = {beta_coeffs_best[0]:.4f} '
coeff = beta_coeffs_best[1]
if coeff >= 0:
    equation_b += f'+ {coeff:.4f} * "{best_feature_b}"'
else:
    equation_b += f'- {-coeff:.4f} * "{best_feature_b}"'
print(equation_b)

print("\nBảng xếp hạng MSE của các đặc trưng (từ thấp đến cao):")
sorted_features = sorted(feature_mse_scores.items(), key=lambda item: item[1])
for feature, mse in sorted_features:
    print(f"- {feature}: {mse:.4f}")

y_true = [row[0] for row in y_data_best]
y_pred = [sum(b * x for b, x in zip(beta_coeffs_best, [1, row[0]])) for row in X_data_best]

plt.figure(figsize=(10,10))
plt.scatter(y_true, y_pred,
    c="royalblue",
    alpha=0.4,        
    edgecolor="white",
    s=40)
plt.plot([min(y_true), max(y_true)], [min(y_true), max(y_true)], 'r--', linewidth=2, label="y = x")

plt.xlabel("Thực tế (Actual)")
plt.ylabel("Dự đoán (Predicted)")
plt.title("So sánh giá trị thực tế và dự đoán")
plt.legend()
plt.show()
  
```

**Nhận xét**: Kết quả khá tương đồng với `Câu a` cho thế sai số khi sử dụng giảm dần ứng với hệ số tương quan tăng dần. Cụ thể là đặc trưng `livarea` có hệ số tương quan cao nhất với `sprice` và cũng là đặc trưng cho mô hình có sai số trung bình `MSE` nhỏ nhất. Mô hình cho thấy xu hướng giá nhà sẽ tăng khi diện tích sử dụng tăng

**Nhận xét:** Đám mây giá trị của `sprice` xuất hiện chủ yếu trong khoảng giá trị từ 50000-250000 và được cắt ngang qua bởi đường thẳng y = x, trong khoảng này mô hình cho thấy xu hướng giá nhà sẽ tăng khi diện tích sử dụng tăng. Ở các giá trị lớn thì dự đoán có sai số lớn, cần thêm các đặc trưng khác để tham chiếu

> **Câu c:** Xây dựng mô hình đa đặc trưng: $sprice = \theta_0 + \theta_1 .X_1 + \theta_2 .X_2 + ...$

```python
df = pd.read_csv('stockton4.csv')
df.columns = df.columns.str.strip()
headers = list(df.columns)
data = df.values.tolist()

feature_cols = [col for col in df.columns if col != 'sprice']
target_col = 'sprice'

X_data = df[feature_cols].values.tolist()
y_data = [[val] for val in df[target_col].values.tolist()]
X_b = [[1.0] + row for row in X_data]

beta = OLS(X_b, y_data)
beta_coeffs = [b[0] for b in beta]
  

        
print("\n--- PHƯƠNG TRÌNH HỒI QUY TUYẾN TÍNH ---")
equation = f"quality = {beta_coeffs[0]:.4f} "
for i, col_name in enumerate(feature_cols):
    coeff = beta_coeffs[i+1]
    if coeff >= 0:
        equation += f"+ {coeff:.4f} * {col_name} \n"
    else:
        equation += f"- {-coeff:.4f} * {col_name} \n"
print(equation)

y_pred = predict(X_b, beta)
mse_c = MSE(y_data, y_pred)
print(f"Trung bình bình phương sai số (MSE): {mse_c:.4f}")

y_true = [row[0] for row in y_data]

plt.figure(figsize=(10,10))
plt.scatter(y_true, y_pred,
    c="royalblue",
    alpha=0.4,        
    edgecolor="white",
    s=40)
plt.plot([min(y_true), max(y_true)], [min(y_true), max(y_true)], 'r--', linewidth=2, label="y = x")

plt.xlabel("Thực tế (Actual)")
plt.ylabel("Dự đoán (Predicted)")
plt.title("So sánh giá trị thực tế và dự đoán:")
plt.legend()
plt.show()
```

**Nhận xét:** Mô hình đã cho thấy xu hướng chung là giá trị `sprice` sẽ tăng khi đặc trưng có tương quan cao nhất là `livarea` tăng với hệ số dương. 2 đặc trưng khác cũng có hệ số tương quan cao là `beds` và `baths` lại mang hệ số âm, có nghĩa khi số lượng phòng tắm và phòng ngủ tăng thì giá nhà sẽ giảm nếu diện tích sử dụng không tăng, cho thấy 2 đặc trưng này cũng có phần phụ thuốc vào đặc trưng `livarea`, diện tích sử dụng tăng thì số phòng ngủ và phòng tắm tăng theo. 3 đặc trưng có hệ số tương quan không quá cao là `lgelot`, `age` và `pool` thì rõ ràng hơn, diện tích đất tăng thì giá cũng tăng, tuổi thọ tăng thì giá giảm, có hồ bơi thì giá tăng

### Bài 2: Dự đoán tình trạng giao thông
> **Câu a:** Tạo dữ liệu mô phỏng

```python
import random

states = ["K", "B", "T"]
n = len(states)

random.seed(2025)
n_days = 60

sim = [random.choice(states) for _ in range(n_days)]
```

> **Câu b:** Xây dựng ma trận đếm số lần chuyển trạng thái 

```python
counts = [[0]*n for _ in range(n)]
idx = {s:i for i,s in enumerate(states)}

for k in range(n_days-1):
    j = idx[sim[k]]     
    i = idx[sim[k+1]]   
    counts[i][j] += 1

print("Ma trận đếm:")
for row in counts:
    print(row)
```

> **Câu c:** Xây dựng ma trận xác suất chuyển trạng thái

```python
P = [[0]*n for _ in range(n)]

for j in range(n):
    col_sum = sum(counts[i][j] for i in range(n))
    if col_sum == 0:
        for i in range(n):
            P[i][j] = 1/n
    else:
        for i in range(n):
            P[i][j] = counts[i][j] / col_sum
print("\nMa trận xác suất chuyển trạng thái P:")
for row in P:
    print([round(x,4) for x in row])
```

> **Câu d:** Dự đoán trạng thái thời tiết

```python
# Hàm tính lũy thừa ma trận P
def prop_distribution(P, k):
    n = len(P)
    mu = [[1.0 / n] for _ in range(n)]
    pk = P
    for _ in range(k - 1):
        pk = matrix_multiply(pk, P)
    
    mu = matrix_multiply(pk, mu)
    
    return mu

# Hàm in phân phối
def print_distribution(mu, states):
    for i, s in enumerate(states):
        print(f"{s}: {mu[i][0]:.4f}")
    print("-" * 30)

print("Phân phối sau 1 ngày:")
print_distribution(prop_distribution(P, 1), states)
print("Phân phối sau 2 ngày:")
print_distribution(prop_distribution(P, 2), states)
print("Phân phối sau 10 ngày:")
print_distribution(prop_distribution(P, 10), states)

```

**Nhận xét:** Dựa vào dữ liệu đã tạo, khả năng xảy ra kẹt xe khá cao ~0.47 cho cả 3 ngày

> **Câu e:** Tìm phân phối dừng

```python
# Hàm giải hệ tuyến tính Ax = b bằng Gauss
def gauss_elimination(A, b, eps=1e-12):
    n = len(A)
    aug = [A[i] + [b[i]] for i in range(n)]
    for col in range(n):
        pivot_row = max(range(col, n), key=lambda r: abs(aug[r][col]))
        if abs(aug[pivot_row][col]) < eps: continue
        aug[col], aug[pivot_row] = aug[pivot_row], aug[col]
        pivot = aug[col][col]
        aug[col] = [val / pivot for val in aug[col]]
        for r in range(n):
            if r != col:
                factor = aug[r][col]
                if abs(factor) > eps:
                    aug[r] = [aug[r][c] - factor*aug[col][c] for c in range(n+1)]
    return [row[-1] for row in aug]

# Hàm tính phân phối dừng
def solve_stationary(P):
    n = len(P)
    PT = transpose_matrix(P)
    
    A = [[PT[i][j] for j in range(n)] for i in range(n)]
    for i in range(n):
        A[i][i] -= 1.0
        
    b = [0.0] * n
    
    A[-1] = [1.0] * n
    b[-1] = 1.0
    
    pi_flat = gauss_elimination(A, b)
    
    s = sum(pi_flat)
    if s > 0:
        pi_normalized = [val / s for val in pi_flat]
    else:
        pi_normalized = pi_flat

    pi_column = [[val] for val in pi_normalized]
    
    return pi_column

pi = solve_stationary(P)

print("Phân phối dừng:")
print_distribution(pi, states)
```

**Nhận xét:** 
- Phân phối dừng cho xác suất cả 3 tình trạng giao thông đều bằng nhau
- Ma trận xác suất chuyển trạng thái P chính quy nên phân phối dừng cũng là phân phối giới hạn cho thấy trong quãng thời gian dài thì các trạng thái sẽ xuất hiện ngẫu nhiên với xác suất 1/3 như nhau

## 3. Giải thích các hàm hoạt động
### Bài 1: Hồi quy tuyến tính
> **3.1 OLS(X,y)**

**Ý tưởng**: Mô phỏng lại công thức  `OLS (Ordinary Least Square)`: $ \theta = (X^{T}.X)^{-1}.X^{T}.Y $

**Mô tả hoạt động**: Sử dụng các hàm tính toán sau cho công thức trên

`transpose_matrix(matrix)`: Tạo ma trận chuyển vị

`inverse_matrix(matrix)`: Tạo ma trận nghịch đảo, dùng thuật toán Gauss-Jordan, các phép biến đổi sơ cấp trên dòng

`matrix_multiply(A, B)`: Phép nhân 2 ma trận


> **3.2 MSE(y_true, y_pred)**

**Ý tưởng**: Mô phỏng lại công thức tính trung bình bình phương sai số `MSE (Medium Square Errors)`: $ \frac{1}{n}\sum_{i=1}^{n}(y_i - ŷ_i)^{2} $

**Mô tả hoạt động**: Dựa vào các giá trị $ ŷ_i $ được trả về từ hàm `predict(X_b,beta)` để tính tổng bình phương sai số `RSS (Residual Sum of Square)` sau đó lấy kết quả vừa tìm chia cho kích thước dữ liệu hay số bộ dữ liệu

> **3.3 k_fold_cross_validation(data, headers, feature_col_names, target_col_name, k=100)**

**Ý tưởng**: Chia tập dữ liệu thành `k` phần, cụ thể trong bài là 100 phần để đánh giá từng đặc trưng, dùng hàm `MSE(y_true, y_pred)` để trả về `MSE`trung bình cho từng đặc trưng

**Mô tả hoạt động**: Chia dữ liệu thành 100 phần mỗi phần có khoảng ~150 hàng. Khi vòng lặp `for` bắt đầu với `k` bằng 100, tập `Train` sẽ lấy lần lượt các phần dữ liệu và các phần còn lại là tập `Test` cho 100 lần, cộng dồn `MSE` lại sau đó đem chia 100 để ra được `MSE` trung bình

### Bài 2: Xích Markov
> **prop_distribution(P, k)**

**Ý tưởng**: Mô phỏng lại `Hệ quả 5` và `Hệ quả 6`: $ \pi_n = P^n.\pi_0 = P^{n-1}.P.\pi_0$

**Mô tả hoạt động**: Dựa vào hàm nhân ma trận `matrix_multiply(A, B)`, tính ma trận $P^n$ sau đó nhân với phân phối đầu $\pi_0$ để tính $\pi_n$

> **gauss_elimination(A, b, eps = 1e-12)**

**Ý tưởng:** Sử dụng thuật toán `Gauss-Jordan` để giải hệ phương trình

**Mô tả hoạt động:**
- Tạo ma trận mở rộng $(A|b)$ 
- Dùng các phép biến đổi trên dòng, mục tiêu biến phần tử trên đường chéo chính thành 1 và các phần tử khác trong cột đó thành 0
- Sau khi khử, phần bên trái của ma trận $(A|b)$ biến thành ma trận đơn vị $I$, phần bên phải hay cột cuối cùng chính là vector nghiệm

> **solve_stationary(P)**

**Ý tưởng**: Sử dụng `Định nghĩa 7` trong phần giới thiệu để xác định phân phối dừng cho xích

**Mô tả hoạt động:**
- Biến đổi công thức $\pi.P = \pi$ về dạng hệ phương trình $Ax = b$ thành $(P-I).\pi = 0$
- Giải hệ này bằng `gauss_elimination()` sẽ thấy kết quả là vô số nghiệm do các phương trình trong hệ có phụ thuộc tuyến tính
- Loại bỏ 1 hàng trong hệ phương trình, thay vào đó là rằng buộc tổng các phần tử của vector phân phối phải bằng 1
- Giải hệ bằng `gauss_elimination()` thu được vector phân phối dừng
